```python
tips = pd.read_csv('tips.csv',
                   skiprows = 1,
                   names = ['Importe_Factura', 'Propina', 'Sexo', 'Fumador', 'Dia',
                              'Tipo', 'Comensales']
                          )
print(len(tips))
tips.head()
```

```python
# Calculo los porcentajes
porcentaje_propina = tips['Propina'].div(tips['Importe_Factura'].astype(float)) * 100
tips['Propina %'] = porcentaje_propina
tips.head()
```

```python
tips['Propina %'].plot(kind = 'hist', 
                       bins = 50, 
                       title = 'Frecuencia de porcentajes');
```

```python
tips['Propina %'].plot(kind = 'kde' , style = 'b');
```

```python
mi_fig, (ax1, ax2) = plt.subplots(nrows= 1,ncols = 2, figsize=(10,5))

tips['Propina %'].plot(kind = 'hist', bins = 50, 
                       ax = ax1);
#---
tips['Propina %'].plot(kind = 'kde' , style = 'b' , 
                       ax = ax2);

```

```python
tips.plot(kind='scatter', x='Importe_Factura', y='Propina');

```

```python
mi_fig, ax1 = plt.subplots(nrows= 1,ncols = 1, figsize=(10,5))


tips[tips.Sexo == 'Male'].plot(kind='scatter', 
                                        x='Importe_Factura', 
                                        y='Propina',
                                        color='DarkBlue',
                                        label = 'Hombres',
                                        ax =  ax1);

tips[tips.Sexo == 'Female'].plot(kind='scatter', 
                                         x='Importe_Factura', 
                                         y='Propina',
                                         color='DarkGreen',
                                         ax = ax1,
                                         label = 'Mujeres');

ax1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title =' Leyenda');
```

```python
comensales_dia = pd.read_excel('comensales_dia.xls')
comensales_dia
```

```python
comensales_dia.Comensales.plot(kind = 'pie',
                              title = 'Numero de comensales por dia',
                              autopct='%.2f');
```
