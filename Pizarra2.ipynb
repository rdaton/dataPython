```python
import numpy as np                         # importación de numpy
import matplotlib as mpl                   # importación de matplotlib con alias mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generación de datos
x = np.linspace(start=0, stop=2*np.pi, num=50)
y1 = np.sin(x)
```

```python
plt.plot(x, y1);
```

```python
# datos
y1 = np.sin(x)
y2 = np.cos(x)  
# representación : 
fig, (ax1, ax2) = plt.subplots(nrows= 1,ncols = 2, figsize=(10,5))
ax1.plot(x, y1)
ax2.plot(x, y2); 
```

```python
fig, ax = plt.subplots(nrows = 1, figsize=(10,5))
ax.plot(x, y1)
ax.plot(x, y2);    # Prueba a quitar el ;
```
